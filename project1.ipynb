# Problem Statement 3: (Rename State/UT Names)
The State/UT names are in all caps in the census data, For uniformity across datasets we use the names so that only the first character of each word in the name is in upper case and the rest are in lower case. However, if the word is “and” then it should be all lowercase.
Examples: 
•	Andaman and Nicobar Islands
•	Arunachal Pradesh
•	Bihar


```python
df3=df2.copy()
df3
```

```python
def func(val):
    if " And " in val[2:-2] or " and " in val[2:-2]:
        val=val.lower()
        v=val.split("and")
        s= str(v[0].strip().title()+" and "+v[1].strip().title())
        return s
    else:
        return val.strip().title()
```

```python
df3["State/UT"]=df3["State/UT"].apply(func)
```

```python
def func_alter(a):
    if "&" in a:
        a1=a.split("&")
        st=str(a1[0].strip().title()+" And "+ a1[1].strip().title())
        return st
    elif a=="Odisha":
        return "Orissa".title()
    elif a=="Puducherry":
        return "Pondicherry".title()
    elif a=="Delhi":
        return "Nct Of Delhi".title()
    else:
        return a.strip().title()
```

```python
def call_procedure(function_name, params):
    connection = engine.raw_connection()
    try:
        cursor = connection.cursor()
        if params != 0:
            cursor.callproc(function_name, [params])
        else:
            cursor.callproc(function_name)
        results = list(cursor.fetchall())
        cursor.close()
        connection.commit()
        return results
    finally:
        connection.close()
```
